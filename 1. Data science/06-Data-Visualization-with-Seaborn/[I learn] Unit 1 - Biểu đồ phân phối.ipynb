```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

# 1. Biểu đồ histogram

Nếu lấy cả biểu đồ rời rạc + liên tục dùng hàm này

```python
sns.histplot(tips["total_bill"], kde=True)
```

Nếu mà cần biểu đò rời rạc không thì:

```python
sns.histplot(tips["total_bill"], kde=False)
```

Nếu mà cần biểu đồ liên tục thì:

```python
sns.kdeplot(tips["total_bill"])
```

Nếu bạn muốn biểu đồ vạch + Biểu đồ đường (còn gọi là biểu đồ Kernel density estimation - KDE)

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

# 2. Biểu đồ scatter

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

# 3. Biểu đồ thể hiện mối tương quan giữa tất cả đối tượng

```python
sns.pairplot(tips)
sns.pairplot(tips, hue='size')
sns.pairplot(tips, hue='sex')
```
