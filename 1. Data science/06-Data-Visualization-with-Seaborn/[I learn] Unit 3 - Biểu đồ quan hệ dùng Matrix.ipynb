```python
import seaborn as sns
```

```python
flights = sns.load_dataset('flights')
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# 1. Biểu đồ heatmap thể hiện mối tương quan (correlation) giữa 2 biến

```python
correction_table = tips[['total_bill', 'tip', 'size']].corr()
```

```python
sns.heatmap(correction_table,cmap='coolwarm',annot=True,linewidths=1)
```

# 2. Biểu đồ heatmap thể hiện số lượng

```python
flights.head()
```

```python
flights_pivot = flights.pivot_table(index='month',columns='year',values='passengers')
flights_pivot
```

```python
sns.heatmap(flights_pivot,cmap='coolwarm',linewidths=1)
```

# 3. Biểu đồ clustermap

```python
sns.clustermap(flights_pivot,cmap='coolwarm',linewidths=1)
```
