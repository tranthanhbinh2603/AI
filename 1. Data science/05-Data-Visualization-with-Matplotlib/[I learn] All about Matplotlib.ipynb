```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
%matplotlib inline
```

```python
x = np.linspace(0, 5, 6)
y = x ** 2
fig = plt.figure(figsize=(3,4), dpi=150)
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, label='x ^ 2', color="#8B008B", alpha=0.5)
ax.plot(y, x, label='sqrt(x)', color="#8B008B")
ax.set_title("title");
ax.set_xlabel("x")
ax.set_ylabel("y");
ax.set_ylim([0, 60])
ax.set_xlim([2, 5])
ax.legend(loc=0) 
fig.savefig("filename.png")
```
